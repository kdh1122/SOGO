```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/WCRC2023/데이터/전처리 중 데이터/CCTV현황_ver2.csv", encoding='euc-kr')
```

```python
df2 = df[["소재지지번주소","위도","경도"]]
```

```python
df2
```

```python
session = df2
for col in session.columns:
    msg = 'column: {:>10}\t Percent of NaN Value: {:.2f}%'.format(col, 100 * (session[col].isnull().sum() / session[col].shape[0]))
    print(msg)
```

```python
df3 = df2.dropna()
print(df3)
```

```python
session = df3
for col in session.columns:
    msg = 'column: {:>10}\t Percent of NaN Value: {:.2f}%'.format(col, 100 * (session[col].isnull().sum() / session[col].shape[0]))
    print(msg)
```

```python
df3
```

```python
# 중복 행 제거 및 CCTV ID 부여
df4 = df3.groupby(['소재지지번주소', '위도', '경도']).size().reset_index(name='count')
df4 = df4.drop_duplicates()

# CCTV ID 부여
df4['cctvid'] = range(1, len(df4) + 1)
df4=df4.drop(columns=['count'])
# 최종 DataFrame 출력
print(df4)
```

```python
from google.colab import files
# df3에서 무작위로 10,000개의 행을 선택하여 df4를 만듭니다.
df5 = df4.sample(n=10000, random_state=42)  # random_state는 랜덤 시드로, 동일한 결과를 얻기 위해 설정합니다.

# df4를 적절한 파일 이름으로 CSV 파일로 저장
df5.to_csv('CCTV현황_ver4.csv', index=False, encoding='euc-kr')
# 파일을 다운로드
files.download('CCTV현황_ver4.csv')
```
