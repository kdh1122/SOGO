```python
import pandas as pd
df1=pd.read_excel("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/측정소_교외대기.xls")
df2=pd.read_excel("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/측정소_국가배경.xls")
df3=pd.read_excel("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/측정소_도로변대기.xls")
df4=pd.read_excel("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/측정소_도시대기.xls")
df5=pd.read_excel("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/측정소_항만.xls")
```

```python
#row0, row 1 drop
df1 = df1.drop([0, 1])
df2 = df2.drop([0, 1])
df3 = df3.drop([0, 1])
df4 = df4.drop([0, 1])
df5 = df5.drop([0, 1])
```

```python
# 데이터프레임들을 concat하여 합치기
result_df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)

# 결과 출력
print(result_df)
```
