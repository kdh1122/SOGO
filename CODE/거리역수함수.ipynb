```python
import pandas as pd

file_path1 = '/content/gdrive/MyDrive/WCRC2023/데이터/최종 데이터/cctv_지하철_중복제거_역수전.csv'
file_path2 = '/content/gdrive/MyDrive/WCRC2023/데이터/최종 데이터/cctv4_학교_중복제거_역수전.csv'

# 파일 불러오기
subway = pd.read_csv(file_path1, encoding='euc-kr')
school = pd.read_csv(file_path2, encoding='euc-kr')

# 데이터 확인
print(subway.head())
print(school.head())
```

```python
# 역수 함수 정의
def reciprocal_weight(distance, epsilon=100):
    return 1 / (distance + epsilon)
```

```python
subway['지하철거리_역수'] = reciprocal_weight(subway['지하철거리'])
print(subway)
```

```python
print(max(subway['지하철거리_역수']))
```

```python
subway['지하철_중요도지수'] = subway['지하철거리_역수'] * subway['지하철이용인원']
subway
```

```python
subway.to_csv('/content/gdrive/MyDrive/WCRC2023/데이터/최종 데이터/cctv_지하철_최종_100.csv', index=False, encoding='euc-kr')
```

```python
# 역수 함수를 적용하여 새로운 열 추가
school['학교거리_역수'] = reciprocal_weight(school['distance'])

# 결과 확인
print(school)
```

```python
school = school.drop(columns=['학교급구분', '설립일자', '설립형태', '본교분교구분', '데이터기준일자', '제공기관코드', '제공기관명', 'feature_x', 'feature_y', 'nearest_x', 'nearest_y'])
school = school.drop(columns=['운영상태', '소재지지번주소_2', '소재지도로명주소', '시도교육청코드', '시도교육청명', '교육지원청코드', '교육지원청명', '생성일자', '변경일자'])
school
```

```python
school = school.rename(columns={'distance': '학교거리'})
```
