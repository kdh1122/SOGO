```python
import pandas as pd

file_path1 = '/content/gdrive/MyDrive/WCRC2023/데이터/최종 데이터/cctv_지하철_최종_10.csv'
file_path2 = '/content/gdrive/MyDrive/WCRC2023/데이터/최종 데이터/cctv_버스300.csv'
file_path3 = '/content/gdrive/MyDrive/WCRC2023/데이터/최종 데이터/cctv_학교_최종_10.csv'

sub = pd.read_csv(file_path1, encoding='euc-kr')
bus = pd.read_csv(file_path2, encoding='euc-kr')
school = pd.read_csv(file_path3, encoding='euc-kr')

print(sub.head())
print(bus.head())
print(school.head())
```

# 지하철 이용인원 시각화

```python
import folium

center_lat, center_lon = sub['위도'].mean(), sub['경도'].mean()

# Folium 지도 객체 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=11)

# 데이터 포인트 추가
for i, row in sub.iterrows():
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=row['지하철이용인원'] / 1000000,  # 지하철이용인원이 클수록 큰 원으로 표시
        color='blue',
        fill=True,
        fill_color='#c634eb',
        fill_opacity=0.6,
        popup=f"{row['역명']} 역\n지하철이용인원: {row['지하철이용인원']}명\n지하철거리: {row['지하철거리']:.2f}m"
    ).add_to(m)

m
```

```python
m.save('/content/gdrive/MyDrive/WCRC2023/시각화/cctv_subway_이용인원.html')
```

```python
from branca.colormap import LinearColormap
import folium

# 지도의 중심을 정하기 위해 평균 위도와 경도 계산
center_lat = sub['위도'].mean()
center_lon = sub['경도'].mean()

# Folium 지도 생성
m = folium.Map(location=[center_lat, center_lon], zoom_start=11)

min_size, max_size = 5, 20

# 지하철이용인원에 따른 그라데이션 색상 설정
colormap = LinearColormap(colors=['#00e676', '#ffea00', '#ff1744'], vmin=sub['지하철이용인원'].min(), vmax=sub['지하철이용인원'].max())

# 데이터 포인트 추가
for i, row in sub.iterrows():
    size = min_size + (max_size - min_size) * (row['지하철이용인원'] - sub['지하철이용인원'].min()) / (sub['지하철이용인원'].max() - sub['지하철이용인원'].min())

    # CircleMarker 테두리 색상 제거
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=size,
        color='',  # 테두리 색상 제거
        fill=True,
        fill_color=colormap(row['지하철이용인원']),
        fill_opacity=0.6,
        popup=f"\ncctvid:{row['cctvid']} \n지하철이용인원: {row['지하철이용인원']:}"
    ).add_to(m)

# Colormap을 맵에 추가
colormap.add_to(m)

m
```

```python
m.save('/content/gdrive/MyDrive/WCRC2023/시각화/cctv_subway_이용인원_ver2.html')
```
