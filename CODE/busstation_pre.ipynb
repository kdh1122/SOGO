```python
import pandas as pd
df10=pd.read_csv("/content/drive/MyDrive/WCRC2023/데이터/전처리 중 데이터/cctv200_버스.csv", encoding='euc-kr')
```

```python
df10
```

```python
nf10 = df10.drop(columns=['위도','경도'])
nf11 = nf10.groupby(['소재지지번주소','위도_2','경도_2']).agg({'이용객수': 'sum', 'fid': 'count'}).reset_index()
```

```python
nf11
```

```python

# "소재지지번주소,위도,경도 로 group후" row 개수를 세고 "이용객수" 열을 합산하여 df11 데이터프레임을 생성
df11 = df10.groupby('소재지지번주소','').agg({'이용객수': 'sum', 'fid': 'count'}).reset_index()
df11.rename(columns={'fid': '버스정류장 수', '이용객수': 'sum(이용객수)'}, inplace=True)

# 결과를 출력
print(df11)
```

```python
from google.colab import files
# df4를 적절한 파일 이름으로 CSV 파일로 저장
df11.to_csv('cctv200_버스_ver2.csv', index=False, encoding='euc-kr')
# 파일을 다운로드
files.download('cctv200_버스_ver2.csv')
```
