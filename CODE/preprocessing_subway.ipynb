```python
file_path1 = '/content/gdrive/MyDrive/WCRC2023/데이터/전처리 중 데이터/지하철_최종_데이터.csv'
file_path2 = '/content/gdrive/MyDrive/WCRC2023/데이터/cctv_ver4 전처리 데이터/CCTV현황_ver4.csv'

# 파일 불러오기
subway_data = pd.read_csv(file_path1, encoding='euc-kr')
cctv_data = pd.read_csv(file_path2, encoding='euc-kr')

# 데이터 확인
print(subway_data.head())
print(cctv_data.head())
```

```python
from scipy.spatial import distance

# Create an empty DataFrame to store the mapping results
mapping_results = pd.DataFrame(columns=['CCTV_ID', 'Nearest_Subway_ID', 'Distance_to_Nearest_Subway'])

# Iterate through CCTV points and find the nearest Subway station
for cctv_index, cctv_row in cctv_data.iterrows():
    cctv_coords = (cctv_row['위도'], cctv_row['경도'])
    nearest_subway_id = None
    min_distance = float('inf')

    for subway_index, subway_row in subway_data.iterrows():
        subway_coords = (subway_row['위도'], subway_row['경도'])
        dist = distance.euclidean(cctv_coords, subway_coords)

        if dist < min_distance:
            min_distance = dist
            nearest_subway_id = subway_row['역ID']

    mapping_results = mapping_results.append({
        'CCTV_ID': cctv_index,
        'Nearest_Subway_ID': nearest_subway_id,
        'Distance_to_Nearest_Subway': min_distance
    }, ignore_index=True)

# Merge the mapping results back to the CCTV DataFrame
cctv_data = pd.concat([cctv_data, mapping_results], axis=1)
```

```python
merged_data = pd.merge(cctv_data, subway_data, left_on='Nearest_Subway_ID', right_on='역ID', how='left')
merged_data
```

```python
merged_data = merged_data.drop(columns=['Nearest_Subway_ID'])
merged_data = merged_data.drop(columns=['철도운영기관명'])
merged_data = merged_data.drop(columns=['선명'])
merged_data = merged_data.drop(columns=['경도_y'])
merged_data = merged_data.drop(columns=['위도_y'])
merged_data = merged_data.drop(columns=['역주소'])
merged_data = merged_data.rename(columns={'위도_x': '위도'})
merged_data = merged_data.rename(columns={'경도_x': '경도'})
merged_data = merged_data.rename(columns={'총합': '지하철역_이용객수'})
merged_data
```

```python
merged_data.to_csv('/content/gdrive/MyDrive/WCRC2023/데이터/전처리 중 데이터/cctv_지하철_최종.csv', index=False, encoding='euc-kr')
```
