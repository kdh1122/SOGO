```python
import pandas as pd
df1=pd.read_csv("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/AWS_DD_2023.csv", encoding='euc-kr')
meta2=pd.read_csv("/content/drive/MyDrive/WCRC2023/데이터/로우 데이터/META_관측지점정보.csv", encoding='euc-kr')
```

```python
# '지점' 및 '지점명' 열을 기준으로 그룹화하고 중복된 값 제거
df11 = df1.groupby(['지점', '지점명']).first().reset_index()
df12=df11[['지점','지점명']]
# 결과를 출력
print(df12)
#현재 운용중인 60개
```

```python
import pandas as pd

# df2와 df12 DataFrame이 이미 생성되어 있다고 가정합니다.

# '지점명' 열을 기준으로 left outer join 수행
merged_df = df12.merge(meta2, on='지점명', how='left')

print(merged_df)
```

```python
import pandas as pd

# Assuming merged_df is your DataFrame
# Filter rows where '종료일' is NaN
filtered_merged_df = merged_df[merged_df['종료일'].isna()]

# The 'filtered_merged_df' DataFrame will contain only rows where '종료일' is NaN.
print(filtered_merged_df)
```

```python
filtered_merged_df2 = filtered_merged_df[filtered_merged_df['지점_x']==(filtered_merged_df['지점_y'])]

# 결과를 출력
print(filtered_merged_df2)
```

```python
row_count = len(filtered_merged_df2)
print("Number of rows in filtered_merged_df2:", row_count)
#59개로 한개가 줄음
#meta 데이터 상에서 종료되었다는 관악(레) 관측소가 종료되지 않음을 확인
final_df=filtered_merged_df2
```

```python
final_df2 = final_df[['지점_x', '지점명', '지점주소', '위도', '경도']]
print(final_df2)
```

```python
new_data = {
    '지점_x': 116,
    '지점명': '관악(레)',
    '지점주소': '경기도 과천시 중앙로 관악산길(관악산기상레이더관측소)',
    '위도': 37.44526,
    '경도': 126.96402
}

final_df3 = final_df2.append(new_data, ignore_index=True)
#관악(레) 관측소 추가
```

```python
print(final_df3)
#60개 운용중인 관측소의 모든 위경도
```

```python
final_df4=final_df3.rename(columns={
    '지점_x': 'AWS_관측지점',
    '지점명': 'AWS_지점명',
    '지점주소': 'AWS_지점주소',
    '위도': 'AWS_위도',
    '경도': 'AWS_경도'
})
```
